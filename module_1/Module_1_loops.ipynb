# Loops

```python
geo_words = ["Shale", "Sandstone", "Limestone", "Granite", "Schist"]
```

```python
print(geo_words[2])
```

Let's make a little loop to print out all of the variables in geo_words

```python
for i in range(len(geo_words)):
  print(geo_words[i])
```
